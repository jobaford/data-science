```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('tennis_stats.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.select_dtypes(include=['float'])
```

```python
corralation = df.drop(['Player'],axis=1)
corralation 
plt.figure(figsize=(16,14))
sns.heatmap(corralation.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.scatter(df['BreakPointsOpportunities'],df['ReturnGamesPlayed'],alpha=0.4)
plt.title('Break Points Opportunities vs Return Games Played')
plt.xlabel('Break Points Opportunities')
plt.ylabel('Return Games Played');
```

```python
x = df[['BreakPointsOpportunities']]
y = df[['ReturnGamesPlayed']]

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=20)
```

```python
ols = LinearRegression()
ols.fit(x_train,y_train)
y_predict = ols.predict(x_test)
```

```python
plt.scatter(y_test,y_predict,alpha=0.4)
plt.title('Actual data vs Predicted data')
plt.xlabel('Actual data (y_test)')
plt.ylabel('Predicted data (y_predict)');
```
